Different ways of creating Pandas
1. From a list
2. from Numpy Array
3. from a dictionary

```python
#from a list
data = [10,20,30,40,50]
s_list = pd.Series(data)
print(s_list)
print(s_list[1])
```

```python
#from a Numpy array
data_array = np.array([10, 20, 30, 40, 50])
s_arr = pd.Series(data_array)
print(s_arr)
print(s_arr[1])
```

```python
#from adictionary
data_dict = {'A': 10,'B': 20,'C': 30,'D': 40,'E': 50}
s_dict = pd.Series(data_dict)
print("Full Series:")
print(s_dict)
```

```python
# Print individual values
print("\nIndividual values:")
for key, value in data_dict.items():
    print(f"Key: {key}, Value: {value}")
```

```python
# Access a single value by key (using [] instead of {})
print("\nAccess value for key 'A':", s_dict['A'])
```

```python
# Access a single value by key (using [] instead of {})
print("\nAccess value for all the keys :", s_dict['A'], s_dict['B'], s_dict['C'], s_dict['D'], s_dict['E'])
```

Dataframe
Dataframes can be created from :
1. List
2. Dictinary
3. CSV file

```python
#1. Dataframe created from a list
#creating a list of lists
data = [[1, 'John', 25],
        [2, 'Alice', 30],
        [3, 'Bob', 45]]

#Creating a Dataframe from the List
df = pd.DataFrame(data, columns=['ID', 'Name', 'Age'])

print(df)
```

```python
# Append new row to the list
data.append([5, 'Julius', 38])

print(df)
```

```python
# Append new row to the list
data.append([4, 'Diana', 28])

# Create DataFrame from updated list
df = pd.DataFrame(data, columns=['ID', 'Name', 'Age'])

print(df)
```

```python
print(data)
```

```python
# Remove the second row (index 1 → Alice)
data.pop(1)

print(data)
```

```python
# Remove Bob’s record directly
data.remove([3, 'Bob', 45])

print(data)
```

DATAFRAME INDEXING OPERATIONS

DATAFRAME INDEXING

```python
# Selecting Columns
print("Selecting Columns:")
print(df['customerID']) # using square brackets
print(df.tenure) # Using dot notation
```

```python
#LOCATION
#selecting rows
print("\nSelecting Rows: ")
print("Using .loc[]: ")
print(df.loc[1]) #label-based indexing
print("Using .iloc[]: ")
print(df.iloc[2]) #Integer-based indexing
print("Using boolean indexing:")
print(df[df['tenure'] > 30]) #booean indexing
```

```python
#conditional selection
print("\conditional selection: ")
print(df[df['gender'] == 'Male']) # selecting rows based on condition
```

```python
#setting index
print("\nSeting Index: ")
df_indexed = df.set_index('customerID')
print(df_indexed)
```
